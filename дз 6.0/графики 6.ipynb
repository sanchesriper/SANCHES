```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
#Задание 1
class NumberChecker:
    def __init__(self):
        self.A = 0
        self.B = 0
        self.C = 0
    def input_numbers(self):
        self.A = int(input("Введите число A: "))
        self.B = int(input("Введите число B: "))
        self.C = int(input("Введите число C: "))
    
    def is_one_even(self):
        return (self.A % 2 == 0) != (self.B % 2 == 0)
    def are_all_multiples_of_three(self):
        return (self.A % 3 == 0) and (self.B % 3 == 0) and (self.C % 3 == 0)
    def visualize_results(self):
        results = {
            'Только одно чётное': self.is_one_even(),
            'Все кратны 3': self.are_all_multiples_of_three()
        }

        labels = list(results.keys())
        values = [int(results) for results in results.values()]

        plt.bar(labels, values, color=['blue', 'green'])
        plt.ylim(0, 1)
        plt.ylabel('Результат(0 - Ложь, 1 - Истина)')
        plt.title('Результат проверки условий')
        plt.show()

checker = NumberChecker()

checker.input_numbers()

print("Только одно из чисел A и B четное:", checker.is_one_even())
print("Каждое из чисел A, B, C кратно трем:", checker.are_all_multiples_of_three())

checker.visualize_results()
```

```python
class NumberChecker:
    def __init__(self):
        self.data = pd.DataFrame(columns=['A', 'B', 'C'])

    def input_numbers(self, a, b, c):
        new_data = pd.DataFrame({'A': [a], 'B': [b], 'C': [c] })
        self.data = pd.concat([self.data, new_data], ignore_index=True)
    
    def is_one_even(self):
        return (self.data['A'] % 2 == 0) != (self.data['B'] % 2 == 0)

    def are_all_multiples_of_three(self):
        return (self.data['A'] % 3 == 0) & (self.data['B'] % 3 == 0) & (self.data['C'] % 3 == 0)

    def visualize_results(self):
        results = {
            'Только одно чётное': self.is_one_even().astype(int),
            'Все кратны 3': self.are_all_multiples_of_three().astype(int)
        }
        
        results_df = pd.DataFrame(results)

        plt.figure(figsize=(10, 5))
        sns.countplot(data=results_df.melt(), x='variable', hue='value', palette='Set1')
        plt.title('Результаты проверки условий')
        plt.ylabel('Количество')
        plt.xlabel('Условия')
        plt.xticks(rotation=45)
        plt.legend(title='Результат', loc='upper right', labels=['Ложь', 'Истина'])
        plt.show()

if __name__ == '__main__':
    checker = NumberChecker()

    A = int(input("Введите число A: "))
    B = int(input("Введите число B: "))
    C = int(input("Введите число C: "))

    checker.input_numbers(A, B, C)

    print("Только одно из чисел A и B четное:", checker.is_one_even().any())
    print("Каждое из чисел A, B, C кратно трем:", checker.are_all_multiples_of_three().any())

    checker.visualize_results()
```

```python
class NumberConditions:
    def __init__(self,start,end):
        self.numbers = pd.DataFrame({'numbers':np.arange(start, end + 1)})

    def check_conditions(self):
        self.numbers['condition_a'] = (self.numbers['numbers'] % 4 == 0) | (self.numbers['numbers'] % 7 == 0)
        self.numbers['condition_b'] = (self.numbers['numbers'] % 5 == 0) | (self.numbers['numbers'] % 10 == 0)

    def get_condition_a(self):
        return self.numbers[self.numbers['condition_a']]
    def get_condition_b(self):
        return self.numbers[self.numbers['condition_b']]
    
    def visual_conditions(self):
        plt.figure(figsize = (12,6))

        plt.subplot(1,2,1)
        sns.scatterplot(data = self.numbers, x = 'numbers', y = 'condition_a', hue='condition_a',palette={True:'green',False:'red'})
        plt.title('Условия а : Кратно 4 или 7')
        plt.xlabel('Числа')
        plt.ylabel('Условие а (истина/ложь)')

        plt.subplot(1,2,2)
        sns.scatterplot(data = self.numbers, x = 'numbers', y = 'condition_b', hue='condition_b',palette={True:'green',False:'red'})
        plt.title('Условия б : Кратно 5 и не оканчивается на 0')
        plt.xlabel('Числа')
        plt.ylabel('Условие b (истина/ложь)')

        plt.tight_layout()
        plt.show()
    
number_conditions = NumberConditions(1,100)
number_conditions.check_conditions()

print("Условие а (кратно 4 или 7)")
print(number_conditions.get_condition_a())

print("\nУсловие b (кратно 5 и не оканчивается на 0)")
print(number_conditions.get_condition_b())

number_conditions.visual_conditions()
```

```python
#Задание 3
class NumberCondition:
    def __init__(self, n, m, k, l):
        self.n = n
        self.m = m
        self.k = k
        self.l = l

    def check_conditions(self):
        condition1 = self.n + self.m > self.k
        if self.n > self.k:
            condition2 = self.m < 1
        else:
            condition2 = True 
        result = condition1 and condition2
        return result

n = int(input('Введите значение n: '))
m = int(input('Введите значение m: '))
k = int(input('Введите значение k: '))
l = int(input('Введите значение l: ')) 

number_condition = NumberCondition(n, m, k, l)
result = number_condition.check_conditions()

if result is not None:
    print("Результат проверки условий:", result)
else:
    print("Условия не выполняются.")

conditions_met = (n + m > k) and ((n <= k) or (m < 1)) 

plt.figure()
plt.title("Logical Conditions Check")
plt.bar(['Conditions Met'], [1 if conditions_met else 0], color='green' if conditions_met else 'red') 
plt.ylim(0, 1.5) 
plt.ylabel("Status")
plt.xticks([])
plt.show()
```

```python
#Задание 6
class MyClassTask:
    def __init__(self, deposit):
        self.deposit = deposit

    def calculate_payout(self):
        if self.deposit < 50000:
            rate = 0.16
        elif self.deposit <= 100000:
            rate = 0.18
        else:
            rate = 0  
        
        return self.deposit * (1 + rate)

deposit = float(input('Введите сумму вклада: '))

my_task = MyClassTask(deposit)
payout = my_task.calculate_payout()

data = pd.DataFrame({"Тип": ["Начальная сумма", "Сумма с процентами"], "Сумма": [deposit, payout]})

plt.figure(figsize=(5, 5))
sns.barplot(data=data, x="Тип", y="Сумма", hue = 'Тип', palette=["blue", "green"])
plt.ylabel("Сумма в рублях")
plt.title("Расчет суммы выплаты по депозиту")
plt.show()

print(f"Сумма выплаты по депозиту: {payout:.2f} руб.")
```

```python
#Задание 14
class LogicExpression:
    def __init__(self):
        self.X = False
        self.Y = False
        self.Z = False

    def input_values(self):
        self.X = bool(int(input("Введите значение X (0 или 1): ")))
        self.Y = bool(int(input("Введите значение Y (0 или 1): ")))
        self.Z = bool(int(input("Введите значение Z (0 или 1): ")))

    def evaluate_expression(self):
        return not (not self.X and self.Y) or (self.X and not self.Z)
    
    def visualize_result(self):
        result = self.evaluate_expression()
        labels = ['Результат выражения']
        values = [int(result)]

        plt.bar(labels, values, color='green')
        plt.ylim(0, 1)
        plt.ylabel('Результат(0 - Ложь, 1 - Истина)')
        plt.title('Значение логического выражения')
        plt.show()

    def generate_truth_table(self):
        print("X | Y | Z | not (not X and Y) or (X and not Z)")
        print("---------------------------------------------")
        for x in [False, True]:
            for y in [False, True]:
                for z in [False, True]:
                    self.X = x
                    self.Y = y
                    self.Z = z
                    result = self.evaluate_expression()
                    print(f"{int(x)} | {int(y)} | {int(z)} | {int(result)}")

logic_checker = LogicExpression()

print("Генерация таблицы истинности:")
logic_checker.generate_truth_table()

print("\nВвод значений переменных вручную:")
logic_checker.input_values()
result = logic_checker.evaluate_expression()
print("Результат выражения:", result)

logic_checker.visualize_result()
```

```python
class NumberChecker:
    def __init__(self):
        self.data = pd.DataFrame(columns=['A', 'B', 'C', 'D'])

    def input_numbers(self, a, b, c, d):
        new_data = pd.DataFrame({'A': [a], 'B': [b], 'C': [c], 'D': [d]})
        self.data = pd.concat([self.data, new_data], ignore_index=True)

    def are_two_even(self):
        a_even = self.data['A'] % 2 == 0
        b_even = self.data['B'] % 2 == 0
        c_even = self.data['C'] % 2 == 0
        d_even = self.data['D'] % 2 == 0

        return (
            (a_even & b_even & ~c_even & ~d_even) |
            (a_even & c_even & ~b_even & ~d_even) |
            (a_even & d_even & ~b_even & ~c_even) |
            (b_even & c_even & ~a_even & ~d_even) |
            (b_even & d_even & ~a_even & ~c_even) |
            (c_even & d_even & ~a_even & ~b_even)
        )

    def check_divisibility(self):
        a_divisible_by_3 = self.data['A'] % 3 == 0
        b_divisible_by_3 = self.data['B'] % 3 == 0
        c_divisible_by_3 = self.data['C'] % 3 == 0
        d_divisible_by_3 = self.data['D'] % 3 == 0

        a_divisible_by_5 = self.data['A'] % 5 == 0
        b_divisible_by_5 = self.data['B'] % 5 == 0
        c_divisible_by_5 = self.data['C'] % 5 == 0
        d_divisible_by_5 = self.data['D'] % 5 == 0

        return (
            (a_divisible_by_3 & b_divisible_by_5 & ~c_divisible_by_3 & ~d_divisible_by_3 & ~a_divisible_by_5 & ~c_divisible_by_5 & ~d_divisible_by_5)
        )


    def is_one_even(self):
        return (self.data['A'] % 2 == 0) != (self.data['B'] % 2 == 0)

    def are_all_multiples_of_three(self):
        return (self.data['A'] % 3 == 0) & (self.data['B'] % 3 == 0) & (self.data['C'] % 3 == 0) & (self.data['D'] % 3 == 0)

    def visualize_results(self):
        results = {
            'есть одно чётное число': self.is_one_even().astype(int),
            '3 четных числа': self.are_all_multiples_of_three().astype(int),
            'Два четных': self.are_two_even().astype(int),
            'Одно на 3, другое на 5': self.check_divisibility().astype(int)
        }

        results_df = pd.DataFrame(results)
        plt.figure(figsize=(12, 6))
        sns.countplot(data=results_df.melt(), x='variable', hue='value', palette='Set1')
        plt.title('Результаты проверки условий')
        plt.ylabel('Количество')
        plt.xlabel('Условия')
        plt.xticks(rotation=45, ha='right')
        plt.legend(title='Результат', loc='upper right', labels=['Ложь', 'Истина'])
        plt.tight_layout()
        plt.show()


if __name__ == '__main__':
    checker = NumberChecker()
    A = int(input("Введите число A: "))
    B = int(input("Введите число B: "))
    C = int(input("Введите число C: "))
    D = int(input("Введите число D: "))
    checker.input_numbers(A, B, C, D)

    print("есть одно четное число:", checker.is_one_even().any())
    print("3 четных числа:", checker.are_all_multiples_of_three().any())
    print("Ровно два из чисел A, B, C и D четные:", checker.are_two_even().any())
    print("Одно делится на 3, другое на 5:", checker.check_divisibility().any())
    checker.visualize_results()
```

```python
#Задание 18
class PurchaseCalculator:
    def __init__(self, purchase_amount):
        self.purchase_amount = purchase_amount

    def calculate_discounted_price(self):
        if 10000 < self.purchase_amount < 20000:
            discount = 0.05
        elif 20000 < self.purchase_amount < 50000:
            discount = 0.10
        else:
            discount = 0.0 

        discounted_price = self.purchase_amount * (1 - discount)
        return discounted_price


purchase_amount = float(input('Введите сумму покупки: '))

calculator = PurchaseCalculator(purchase_amount)
discounted_price = calculator.calculate_discounted_price()

print(f"Итоговая стоимость покупки со скидкой: {discounted_price:.2f} руб.")

data = pd.DataFrame({
    "Тип": ["Начальная сумма", "Сумма со скидкой"],
    "Сумма": [purchase_amount, discounted_price]
})

plt.figure(figsize=(5, 5))
sns.barplot(data=data, x="Тип", y="Сумма", hue='Тип', palette=["blue", "green"])
plt.ylabel("Сумма в рублях")
plt.title("Расчет стоимости покупки со скидкой")
plt.show()
```

```python
#Задание 20
class LogicExpression:
    def __init__(self):
        self.X = False
        self.Y = False
        self.Z = False

    def input_values(self):
        self.X = bool(int(input("Введите значение X (0 или 1): ")))
        self.Y = bool(int(input("Введите значение Y (0 или 1): ")))
        self.Z = bool(int(input("Введите значение Z (0 или 1): ")))

    def evaluate_expression(self):
        return (self.X and not (not self.Y or self.Z)) or self.Y

    def visualize_result(self):
        result = self.evaluate_expression()
        labels = ['Результат выражения']
        values = [int(result)]
        plt.bar(labels, values, color='green')
        plt.ylim(0, 1)
        plt.ylabel('Результат (0 - Ложь, 1 - Истина)')
        plt.title('Значение логического выражения')
        plt.show()

    def generate_truth_table(self):
        print("X | Y | Z | X and not (not Y or Z) or Y")
        print("---------------------------------------------")
        for x in [False, True]:
            for y in [False, True]:
                for z in [False, True]:
                    self.X = x
                    self.Y = y
                    self.Z = z
                    result = self.evaluate_expression()
                    print(f"{int(x)} | {int(y)} | {int(z)} | {int(result)}")

logic_checker = LogicExpression()

print("Генерация таблицы истинности:")
logic_checker.generate_truth_table()

print("\nВвод значений переменных вручную:")
logic_checker.input_values()
result = logic_checker.evaluate_expression()
print("Результат выражения:", result)
logic_checker.visualize_result()
```

```python
#Задание 21
class ExperienceCoefficient:
    def __init__(self, experience):
        self.experience = experience

    def calculate_coefficient(self):
        if self.experience < 2:
            coefficient = 11
        elif 2 <= self.experience <= 5:
            coefficient = 12
        else:
            coefficient = 13
        return coefficient

experience = float(input('Введите стаж работы по специальности (в годах): '))
coefficient_calculator = ExperienceCoefficient(experience)
coefficient = coefficient_calculator.calculate_coefficient()
print(f"Коэффициент учета стажа: {coefficient}")

data = {'Стаж': [experience], 'Коэффициент': [coefficient]}
df = pd.DataFrame(data)

plt.figure(figsize=(6, 4))
sns.barplot(x='Стаж', y='Коэффициент', data=df)
plt.title('Коэффициент учета стажа работы')
plt.xlabel('Стаж работы (годы)')
plt.ylabel('Коэффициент')
plt.show()
```
