```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Задание 1

```python
a = float(input('Введите значение для переменной a: '))
c = float(input('Введите значение для переменной c: '))

x_values = np.linspace(-10, 10, 400)

L_values = np.sqrt(np.exp(x_values) - 4 * np.cos(x_values**2 * a **5))  + 4 * np.tan(a - x_values**5) / np.e * np.sqrt(np.abs(a + x_values * c**4))

plt.figure(figsize=(10, 6))
plt.plot(x_values, L_values, label='L(x)', color='blue')
plt.title('График функции L(x)')
plt.xlabel('x')
plt.ylabel('L')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.ylim(-10, 10)
plt.show()
```
