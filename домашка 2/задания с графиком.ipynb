```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Задание 1
G = np.e * 2 * y + np.sin(f) / np.log10(3.8 * y + f)

y_values = np.linspace(0.1, 10, 100)
f = 0
G_values = np.e * 2 * y_values + np.sin(f) / np.log10(3.8 * y_values + f)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x=y_values, y=G_values, label='G(y)', color='blue')

plt.xlabel('y')
plt.ylabel('G')
plt.title('График G как функция от y')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
