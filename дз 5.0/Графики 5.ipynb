```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Задание 1
class LogicalExpressions:
    def __init__(self, A, B, C):
        self.A = A
        self.B = B
        self.C = C

    def expression_a(self):
        return (self.A or (not self.A and self.B)) or self.C

    def expression_b(self):
        return (not self.A) or (self.A and (self.B or self.C))

    def expression_c(self):
        return (self.A or (self.B and not self.C)) and self.C


values = [(A, B, C) for A in [True, False] for B in [True, False] for C in [True, False]]


results_a = []
results_b = []
results_c = []


for A, B, C in values:
    logical_expressions = LogicalExpressions(A, B, C)
    results_a.append(logical_expressions.expression_a())
    results_b.append(logical_expressions.expression_b())
    results_c.append(logical_expressions.expression_c())


x_labels = [f"A={int(A)}, B={int(B)}, C={int(C)}" for A, B, C in values]
x = np.arange(len(x_labels))


plt.figure(figsize=(12, 6))
plt.plot(x, results_a, marker='o', label='Результат a', linestyle='-')
plt.plot(x, results_b, marker='o', label='Результат б', linestyle='-')
plt.plot(x, results_c, marker='o', label='Результат в', linestyle='-')


plt.xticks(x, x_labels, rotation=45)
plt.ylim(-0.1, 1.1)
plt.axhline(0.5, color='gray', linestyle='--', linewidth=0.7)
plt.title('Результаты логических выражений')
plt.xlabel('Комбинации A, B, C')
plt.ylabel('Результат (True/False)')
plt.legend()
plt.grid()
plt.tight_layout()

plt.show()
```

```python
#Задание 3
class LogicalExpressions:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def expression_a(self):
        return (not (self.x * self.y < 0)) and (self.y > self.x)

    def expression_b(self):
        return (self.x >= 2) or (self.y**2 != 5)

values = [(x, y) for x in range(1, 4) for y in range(-3, 4)]

results_a = []
results_b = []

for x, y in values:
    logical_expressions = LogicalExpressions(x, y)
    results_a.append(logical_expressions.expression_a())
    results_b.append(logical_expressions.expression_b())

x_labels = [f"x={x}, y={y}" for x, y in values]
x = np.arange(len(x_labels))

plt.figure(figsize=(12, 6))
plt.plot(x, results_a, marker='o', label='Результат a', linestyle='-')
plt.plot(x, results_b, marker='o', label='Результат b', linestyle='-')

plt.xticks(x, x_labels, rotation=45)
plt.ylim(-0.1, 1.1)
plt.axhline(0.5, color='gray', linestyle='--', linewidth=0.7)
plt.title('Результаты логических выражений')
plt.xlabel('Комбинации x, y')
plt.ylabel('Результат (True/False)')
plt.legend()
plt.grid()
plt.tight_layout()

plt.show()
```
